```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy
```

First I'll load the data on the binding energies. The data will be stored in a list of dictionaries with the following keys:

   Key   | Value 
---------|----------------
 `'Z'`   | $Z$
 `'A'`   | $Z + N$
 `'BE'`  | Binding Energy
 `'err'` | Error in BE
 `'j'`   | (Not used)
 `'N'`   | $N$

```python
bedata = []
keys = ('Z', 'A', 'BE', 'err', 'j', 'N')
with open('bedata.dat') as f:
    for line in f:
        istr = line.rstrip().split()
        items = map(lambda x: float(x) if '.' in x else int(x), istr)
        items_dict = {k: v for k, v in zip(keys, list(items))}
        bedata.append(items_dict)
```

## Part 1: One-neutron separation energies

The dictionary below will be helpful for organizing my results and labeling the plot.

```python
elmnts = {'O': 8,
          'Ca': 20,
          'Ni': 28,
          'Sn': 50,
          'Pb': 82}
```

### Experimental Data

Next, I'll find the separation energies for each element using the provided formula:

$$S_n = BE(N,Z) - BE(N-1,Z)$$

```python
n_sep_data = {}
for k, v in elmnts.items():
    edata = {j['N']: j['BE'] for j in bedata if j['Z'] == v}
    nsdata = []
    for n in edata.keys():
        try:
            # Calculate the separation energy
            nsdata.append((n, edata[n] - edata[n-1]))
        except KeyError as e:
            # Expect this to be raised for the first data point
            pass
    nsdata = numpy.array(nsdata)  # using numpy makes plotting easier
    nsdata = nsdata[nsdata[:,0].argsort()]  # sorts the array by the first column
    n_sep_data[k] = nsdata
```

### Liquid Drop Model Data

For comparison, I'll find the values from the liquid drop model. This uses the formula below:

$$ BE(N,Z) = \alpha_1 A - \alpha_2 A^{2/3} - \alpha_3 \frac{Z^2}{A^{1/3}} - \alpha_4 \frac{(Z-N)^2}{A} $$

The values of the $\alpha_n$ are given below in the function definition.

```python
def ldrop_be(N, Z):
    a1 = 15.49
    a2 = 17.23
    a3 = 0.697
    a4 = 22.6
    
    A = N + Z
    
    return a1*A - a2*A**(2/3) - a3*Z**2/A**(1/3) - a4*(Z-N)**2/A
```

Now I'll find the binding energies and one-neutron separation energies, as above.

```python
ld_bes = {}  # The binding energies
for label, z in elmnts.items():
    ns = n_sep_data[label][:,0]
    bes = ldrop_be(z, ns)
    ld_bes[label] = numpy.vstack((ns, bes)).T
    
ld_n_sep = {}  # The separation energies
for label, data in ld_bes.items():
    ens = numpy.diff(data, axis=0)[:,1]
    ld_n_sep[label] = numpy.vstack((data[1:,0], ens)).T
```

```python
fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(15)
nrows = 3
ncols = 2
assert len(elmnts) < nrows * ncols
for idx, lbl in enumerate(sorted(elmnts, key=elmnts.get)):
    plt.subplot(nrows, ncols, idx+1)
    expt = n_sep_data[lbl]
    ld = ld_n_sep[lbl]
    plt.plot(expt[:,0], expt[:,1], '.-', label='Experiment')
    plt.plot(ld[:,0], ld[:,1], '--', label='Liquid Drop')
    plt.legend(loc='best')
    plt.title(lbl + ' Isotope Neutron Separation Energies')
    plt.xlabel('$N$')
    plt.ylabel('$S_n$ [MeV]')
```

Based on these plots, it appears that the liquid drop model does not accurately predict the separation energies of the heavier nuclei.
