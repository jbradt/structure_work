```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy
```

First I'll load the data on the binding energies. The data will be stored in a list of dictionaries with the following keys:

   Key   | Value 
---------|----------------
 `'Z'`   | $Z$
 `'A'`   | $Z + N$
 `'BE'`  | Binding Energy
 `'err'` | Error in BE
 `'j'`   | (Not used)
 `'N'`   | $N$

```python
bedata = []
keys = ('Z', 'A', 'BE', 'err', 'j', 'N')
with open('bedata.dat') as f:
    for line in f:
        istr = line.rstrip().split()
        items = map(lambda x: float(x) if '.' in x else int(x), istr)
        items_dict = {k: v for k, v in zip(keys, list(items))}
        bedata.append(items_dict)
```

## Part 1: One-neutron separation energies

The dictionary below will be helpful for organizing my results and labeling the plot.

```python
elmnts = {'O': 8,
          'Ca': 20,
          'Ni': 28,
          'Sn': 50,
          'Pb': 82}
```

### Experimental Data

Next, I'll find the separation energies for each element using the provided formula:

$$S_n = BE(N,Z) - BE(N-1,Z)$$

```python
n_sep_data = {}
for label, z in elmnts.items():
    data = {j['N']: j['BE'] for j in bedata if j['Z'] == z}
    nsep = []
    for n in data.keys():
        try:
            nsep.append([n, data[n] - data[n-1]])
        except KeyError:
            # This is expected since not all N are present
            pass
    nsep = numpy.array(nsep)
    nsep = nsep[nsep[:,0].argsort()]  # sort by n
    n_sep_data[label] = nsep
```

### Liquid Drop Model Data

For comparison, I'll find the values from the liquid drop model. This uses the formula below:

$$ BE(N,Z) = \alpha_1 A - \alpha_2 A^{2/3} - \alpha_3 \frac{Z^2}{A^{1/3}} - \alpha_4 \frac{(Z-N)^2}{A} $$

The values of the $\alpha_n$ are given below in the function definition.

```python
def ldrop_be(N, Z, a1=15.49, a2=17.23, a3=0.697, a4=22.6):
    
    A = N + Z
    
    return a1*A - a2*A**(2/3) - a3*Z**2/A**(1/3) - a4*(N-Z)**2/A
```

Now I'll find the binding energies and one-neutron separation energies, as above.

```python
ld_n_sep = {}  # The separation energies
for label, z in elmnts.items():
    exp_ns = n_sep_data[label][:,0]
    ns = numpy.arange(exp_ns.min(), exp_ns.max()+1)
    nseps = ldrop_be(ns, z) - ldrop_be(ns - 1, z)
    ld_n_sep[label] = numpy.vstack((ns, nseps)).T
```

```python
fig = plt.figure()
fig.set_figwidth(15)
fig.set_figheight(15)
nrows = 3
ncols = 2
assert len(elmnts) < nrows * ncols
for idx, lbl in enumerate(sorted(elmnts, key=elmnts.get)):
    ax = plt.subplot(nrows, ncols, idx+1)
    expt = n_sep_data[lbl]
    ld = ld_n_sep[lbl]
    plt.plot(expt[:,0], expt[:,1], '.-', label='Experiment')
    plt.plot(ld[:,0], ld[:,1], '--', label='Liquid Drop')
    plt.legend(loc='best')
    plt.title(lbl + ' Isotope Neutron Separation Energies')
    plt.xlabel('$N$')
    plt.ylabel('$S_n$ [MeV]')
    for x in (2, 8, 20, 28, 50, 82, 126):
        ax.axvline(x, linestyle='-', color='gray', linewidth=1.5)
    plt.xlim(expt[:,0].min(), expt[:,0].max())
```

These plots show that the liquid drop model does a decent job of predicting the neutron separation energies of these elements. That being said, there are a few features that it doesn't explain:

1. There are differences in binding energy between nuclei with even and odd numbers of neutrons. As Alex Brown states in his notes, nuclei with even numbers of nucleons are more bound than odd nuclei. This could be fixed by adding another correction to the liquid drop model BE equation we used.

2. There are large drops in each graph after a magic number (marked by the heavy vertical gray lines). This happens because a new neutron shell opens after each magic number, causing the newly added neutrons to be less bound than those in the closed shell.

## Part 2: Binding Energies vs. Liquid Drop Model

Below, I've plotted the binding energies as a function of $A$. There are two plots. The first one is experimental data from the `bedata.dat` file, and the second one is from the liquid drop model. The liquid drop model data is plotted using the same $(N,Z)$ pairs as in the experimental data.

```python
be_vs_a = numpy.array([[d['A'], d['BE']] for d in bedata])

nzpairs = [(d['N'], d['Z']) for d in bedata]

ld_be_vs_a = numpy.zeros((len(nzpairs), 2))
for idx, p in enumerate(nzpairs):
    ld_be_vs_a[idx] = (p[0] + p[1], ldrop_be(*p))
```

```python
fig = plt.figure()
fig.set_figwidth(15)
plt.subplot(121)
plt.plot(be_vs_a[:,0], be_vs_a[:,1], '.')
plt.xlabel('$A$')
plt.ylabel('$BE(N,Z)$')
plt.ylim((-100,2000))
plt.title('Experimental Binding Energies vs. $A$')
plt.subplot(122)
plt.plot(ld_be_vs_a[:,0], ld_be_vs_a[:,1], '.')
plt.xlabel('$A$')
plt.ylabel('$BE(N,Z)$')
plt.ylim((-100,2000))
plt.title('Liquid Drop Model Binding Energies vs. $A$')
```

These two graphs are nearly identical, though there is a bit of a difference for the lowest values of $A$, where the liquid drop model appears to predict that some nuclei should be unbound.

Next, we can look at the behavior of the liquid drop model with respect to its parameters. As in Alex Brown's Fig. 2.5, I'll start with just the volume term, and then I'll add in one more term at a time to see the behavior change.

```python
zeroed_data = numpy.zeros((4, len(nzpairs), 2))
for idx, p in enumerate(nzpairs):
    x = p[0] + p[1]
    zeroed_data[3, idx] = (x, ldrop_be(*p) / x)
    zeroed_data[2, idx] = (x, ldrop_be(*p, a4=0) / x)
    zeroed_data[1, idx] = (x, ldrop_be(*p, a3=0, a4=0) / x)
    zeroed_data[0, idx] = (x, ldrop_be(*p, a2=0, a3=0, a4=0) / x)
```

```python
for idx, data in enumerate(zeroed_data):
    plt.plot(data[:,0], data[:,1], '.')
plt.xlabel('A')
plt.ylabel('BE/A [MeV]')
plt.legend(('V', 'V + S', 'V + S + C', 'Full'), loc='lower right')
plt.title('Comparison of Liquid Drop Model Terms')
```

In the legend of the plot above, "V", "S", and "C" represent the volume, surface, and Coulomb terms, respectively.
